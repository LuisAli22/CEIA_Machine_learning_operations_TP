# 1 Análisis de datos

## 1.1 Introducción

### 1.1.1 Introducción al Problema
El trabajo práctico consiste en entrenar y comparar modelos de Machine Learning para predecir la **Masa Invariante ($M$)** de una partícula madre a partir de los datos cinemáticos de sus productos de desintegración (dos electrones). 

El dataset utilizado contiene información sobre la energía, los momentos y las coordenadas angulares de ambos electrones detectados tras una colisión.

### 1.1.2 ¿Por qué predecir la Masa Invariante?
Aunque existe una fórmula física determinística basada en la relatividad especial:

$$M = \sqrt{(E_1 + E_2)^2 - \|\vec{p}_1 + \vec{p}_2\|^2}$$



el uso de modelos de aprendizaje supervisado en este contexto es fundamental por:
* **Robustez ante el ruido:** Los detectores tienen errores de medición. Los modelos de ML pueden aprender a filtrar o compensar este ruido.
* **Velocidad de procesamiento:** En sistemas de *trigger* (decisión rápida) del LHC, modelos optimizados pueden identificar partículas de interés en tiempo real antes de realizar cálculos cinemáticos pesados.
* **Evaluación de modelos:** Al tener una respuesta teórica exacta, podemos medir con precisión la capacidad de los algoritmos para aprender relaciones no lineales (raíces y potencias).

### 1.1.3 Metodología: Modelos a Utilizar
Para este problema de regresión, evaluaremos cuatro enfoques con distintos niveles de complejidad:

1. **Regresión Lineal:** Actuará como nuestro *baseline*. Dado que la física del problema no es lineal, nos servirá para cuantificar la mejora de los modelos complejos.
2. **Decision Tree Regressor:** Capaz de capturar relaciones no lineales mediante particiones sucesivas del espacio de datos.
3. **Random Forest:** Un modelo de ensamble (*bagging*) que combina múltiples árboles para reducir la varianza y evitar el sobreajuste.
4. **XGBoost:** Algoritmo de *Gradient Boosting* de alto rendimiento, optimizado para manejar grandes volúmenes de datos y capturar patrones complejos con gran precisión.



### 1.1.4 Objetivo del Trabajo
Identificar cuál de los modelos logra la mejor aproximación a la masa invariante real, utilizando métricas de error como el MSE (Error Cuadrático Medio) y el coeficiente de determinación $R^2$.

### 1.3.3 Ordenamiento de filas

Justificación: Reordenamiento de Variables por $p_T$

Se intercambian en bloque todas las propiedades de los electrones para asegurar que el **Electrón 1** sea siempre el de mayor momento transversal ($p_{T1} \geq p_{T2}$).

* **Integridad Física:** Se mueven todas las columnas en conjunto para mantener la coherencia de cada partícula y no mezclar datos.
* **Generalización:** Al eliminar la ambigüedad del orden, el modelo no desperdicia capacidad aprendiendo simetrías y se vuelve más preciso al reducir el espacio de búsqueda.

```python
# 2. Creamos la máscara: True donde el electrón 2 es más rápido que el 1
mask_swap = df['pt1'] < df['pt2']

# 3. Definimos todas las parejas de columnas que hay que dar vuelta
# Si no rotás todas, terminás con un electrón "Frankenstein" (energía de uno y carga de otro)
pairs_to_swap = [
    ('E1', 'E2'),
    ('px1', 'px2'),
    ('py1', 'py2'),
    ('pz1', 'pz2'),
    ('pt1', 'pt2'),
    ('eta1', 'eta2'),
    ('phi1', 'phi2'),
    ('Q1', 'Q2')
]

# 4. Ejecutamos el intercambio solo en las filas donde la máscara es True
for col1, col2 in pairs_to_swap:
    # Usamos .values para evitar que Pandas intente alinear los nombres de las columnas
    df.loc[mask_swap, [col1, col2]] = df.loc[mask_swap, [col2, col1]].values

# 5. Verificación final
reorder_count = np.where(df['pt1'] >= df['pt2'], 1, 0).sum()

explicacion_tecnica = rf"""
#### Análisis de Distribución Cinemática ($p_T$)
**Validación de Simetría y Ordenamiento Leading/Sub-leading**

Tras el proceso de limpieza y reordenamiento, el dataset presenta una estructura consistente para el entrenamiento del modelo:

* **Identificación de Partículas:** Se ha garantizado que el **Electrón 1** sea siempre el de mayor momento transversal ($pt_1 \geq pt_2$). En física de colisiones, esto se conoce como el electrón *Leading*, mientras que el segundo es el *Sub-leading*."""
if len(df) == reorder_count:
    parte_consistencia = rf""""
* **Consistencia del Dataset:** De los **{len(df):,}** eventos procesados, el 100% cumple con la jerarquía de energía. Esto reduce la varianza del modelo al eliminar la ambigüedad en las etiquetas de entrada."""
else:
    parte_consistencia = rf"""
* **Consistencia del Dataset:** De los **{len(df):,}** eventos procesados, el ***{reorder_count*100/len(df)} ***% cumple con la jerarquía de energía.
    """
parte_comportamiento_fisico = rf"""
* **Comportamiento Físico:** Como se observa en el histograma, la curva azul (Leading) domina las energías altas, mientras que la dorada (Sub-leading) se concentra en la zona de bajo $p_T$, lo cual es el comportamiento esperado en el decaimiento de un Bosón.
"""

display(Markdown(explicacion_tecnica+parte_consistencia+parte_comportamiento_fisico))

display(HTML(rf"""
<div style="background-color: #1e272e; padding: 10px; border-radius: 8px; border-left: 6px solid #3498db; color: white;">
    <span style="font-size: 1.2em;">🔍 <b>Verificación de Integridad:</b></span>
    <span style="margin-left: 20px; color: #2ecc71;">✅ {len(df):,} eventos ordenados correctamente.</span>
</div>
<br>
"""))
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Graficamos ambos momentos transversales
sns.histplot(df['pt1'], color="skyblue", label="Electrón 1 (Leading)", kde=True, stat="density", alpha=0.6)
sns.histplot(df['pt2'], color="orange", label="Electrón 2 (Sub-leading)", kde=True, stat="density", alpha=0.6)

plt.title("Distribución de Momento Transversal ($p_T$) post-ordenamiento", fontsize=14)
plt.xlabel("Momentum Transversal [GeV]", fontsize=12)
plt.ylabel("Densidad", fontsize=12)
plt.legend()
plt.xlim(0, 60) # Ajustado según los datos típicos del CERN

plt.show()
```

### 1.3.6 Validación del Ángulo Azimutal ($\phi$)

Se verificó que los valores de $\phi$ se encuentren dentro del rango estándar $[-\pi, \pi]$ radianes.

## 1.4 Feature Engineering

Para optimizar la capacidad de los modelos de Machine Learning en la reconstrucción de la masa invariante, se realizó la siguiente ingeniería de variables:

1. Delta Phi (Δφ) y Corrección Circular
* **Justificación:** Es el discriminador principal de señales de resonancia. En decaimientos de partículas pesadas (como el Bosón Z), los productos hijos tienden a salir en direcciones opuestas en el plano transversal (Δφ ≈ π).
* **Tratamiento:** Se aplicó una corrección para asegurar que la distancia angular se mantenga en el rango [0, π], eliminando la discontinuidad del detector cilíndrico.

2. Delta Eta (Δη) y Delta R (ΔR)
* **Justificación:** Estas variables definen la topología 3D del evento. ΔR permite separar la "señal" (electrones aislados) del "ruido" (jets de partículas muy cercanos entre sí).
* **Impacto:** Ayuda a los modelos a filtrar el fondo de baja energía que se acumula cerca de 0 y 15 GeV.

3. Producto y Ratio de Momentos (pt_product / pt_ratio)
* **Justificación:** La fórmula física de la masa depende del producto de los momentos transversales. Al entregar esta variable ya calculada, se reduce la complejidad no lineal que el modelo debe aprender, mejorando la precisión en modelos de regresión lineal y redes neuronales.

4. Is_OS (Cargas Opuestas)
* **Justificación:** Actúa como un filtro de conservación de carga. Identifica eventos físicamente posibles (hijos de una madre neutra) frente al ruido combinatorio o "fondo de Maxwell-Boltzmann" generado por coincidencias aleatorias (cargas iguales).

```python
def generar_columnas_fisicas(df):
    # 1. Diferencias angulares
    # Delta Eta: Separación longitudinal
    df['delta_eta'] = (df['eta1'] - df['eta2']).abs()
    
    # Delta Phi: Separación en el plano transversal con corrección de contorno
    # Esto asegura que si una partícula está en 3.1 y la otra en -3.1, la distancia sea 0.1 y no 6.2
    df['delta_phi'] = np.abs(df['phi1'] - df['phi2'])
    df['delta_phi'] = np.where(df['delta_phi'] > np.pi, 2*np.pi - df['delta_phi'], df['delta_phi'])
    
    # 2. Delta R: Distancia angular total (el "cono" de separación)
    df['delta_R'] = np.sqrt(df['delta_eta']**2 + df['delta_phi']**2)
    
    # 3. Combinaciones cinemáticas (ayudan a modelos lineales y redes neuronales)
    df['pt_product'] = df['pt1'] * df['pt2']
    df['pt_ratio'] = df['pt1'] / df['pt2'] # Proporción de simetría del evento
    df['E_total'] = df['E1'] + df['E2']     # Escala de energía total
    
    # 4. Variable de carga (Filtro de señal vs ruido)
    # 1 si son opuestas (posible señal), 0 si son iguales (puro ruido/fondo)
    df['is_os'] = (df['Q1'] != df['Q2']).astype(int)
    
    return df

# Aplicar la función
df = generar_columnas_fisicas(df)
```
